```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
import os

df = pd.read_csv('V2M-20k.txt', header=None, names=['id'])

video_folder = os.path.join('data', 'video')
video_ids = set()
video_ext = set()
for entry in os.scandir(video_folder):
    if entry.is_file():
        name, ext = os.path.splitext(entry.name)
        video_ids.add(name)
        video_ext.add(ext)

print("Video extensions:", video_ext)
df['video exists'] = (df['id'].astype(str)).isin(video_ids)

audio_folder = os.path.join('data', 'audio')

audio_ids = set()
audio_ext = set()
for entry in os.scandir(audio_folder):
    if entry.is_file():
        name, ext = os.path.splitext(entry.name)
        audio_ids.add(name)
        audio_ext.add(ext)

print("Audio extensions:", audio_ext)
df['audio exists'] = (df['id'].astype(str)).isin(audio_ids)

#missing_df = df[(~df['video exists']) ^ (~df['audio exists'])]
missing_df = df[(~df['video exists']) | (~df['audio exists'])]
not_downloaded_df = df[(~df['video exists']) & (~df['audio exists'])]
no_audio_df = df[(df['video exists']) & (~df['audio exists'])]
no_video_df = df[(~df['video exists']) & (df['audio exists'])]

target = no_audio_df
for i in range(len(target)):
    video_url = f"https://www.youtube.com/watch?v={no_video_df['id'].values[i]}"
    print(f"Video {i} URL :", video_url)

print("Downloaded Videos and Audios:", df.shape[0]-missing_df.shape[0])
print("Missing Audios:", no_audio_df.shape[0])
print("Missing Videos", no_video_df.shape[0])
print("Missing Download:", not_downloaded_df.shape[0])
#display(not_downloaded_df)
display(no_audio_df)
```
