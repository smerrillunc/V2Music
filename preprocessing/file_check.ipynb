```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
import os

df = pd.read_csv('V2M-20k.txt', header=None, names=['id'])

video_folder = os.path.join('data', 'video')
video_set = set(os.listdir(video_folder))
df['video exists'] = (df['id'].astype(str) + ".mp4").isin(video_set)

audio_folder = os.path.join('data', 'audio')
audio_set = set(os.listdir(audio_folder))
df['audio exists'] = (df['id'].astype(str) + ".webm").isin(audio_set)

#missing_df = df[(~df['video exists']) ^ (~df['audio exists'])]
missing_df = df[(~df['video exists']) | (~df['audio exists'])]
not_downloaded_df = df[(~df['video exists']) & (~df['audio exists'])]
no_audio_df = df[(df['video exists']) & (~df['audio exists'])]
no_video_df = df[(~df['video exists']) & (df['audio exists'])]


video_url = f"https://www.youtube.com/watch?v={no_audio_df['id'].values[20]}"
print("Video URL:", video_url)

print("Downloaded Videos and Audios:", df.shape[0]-missing_df.shape[0])
print("Missing Audios:", no_audio_df.shape[0])
print("Missing Videos", no_video_df.shape[0])
print("Missing Download:", not_downloaded_df.shape[0])
display(not_downloaded_df)
#display(no_audio_df)
```
